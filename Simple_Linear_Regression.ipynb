Linear Regression




```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
candy = pd.read_csv("https://raw.githubusercontent.com/MilanBinsMathew/Simple_Linear_Regression/main/candy-data.csv")
candy.head()
```

```python
candy.shape
candy.info()
candy.describe()
print(candy.isnull())
candy.plot(x = 'competitorname', y = 'sugarpercent',figsize = (20,5))
```

Single Linear Regression

```python
model = LinearRegression()
```

```python
x = candy['sugarpercent']
x = x.values.reshape(-1,1)
y = candy['winpercent']
model.fit(x,y)
```

```python
print("Coefficient of Determination: {}".format(model.score(x,y)))
print("Intercept: {}".format(model.intercept_))
print("Slope: {}".format(model.coef_))
```

```python
x = [[0.777]]
y_pred = model.predict(x)
print("Predicted response for given input = {}".format(y_pred))
```

Multiple Linear Regression

```python
x = candy[['sugarpercent','pricepercent']]
model = LinearRegression().fit(x,y)
```

```python
print("Coefficient of Determination: {}".format(model.score(x,y)))
print("Intercept: {}".format(model.intercept_))
print("Slope: {}".format(model.coef_))
```

```python

y_pred = model.predict([[0.604,0.837]])
print('Predicted Response = {}'.format(y_pred))
```
